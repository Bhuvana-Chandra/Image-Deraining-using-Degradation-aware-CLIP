```python
# Initialize a dictionary to store the number of parameters for each layer
num_params_per_layer = {}
# Initialize a variable to store the total number of parameters
total_params = 0
import torch

# Load the model from .pth file
model_path = '/kaggle/input/da_clip_derain_batchsize6_epochs133/pytorch/da_clip_derain_batchsize6_epochs133/1/40000_G.pth'
state_dict = torch.load(model_path)

# Calculate the number of parameters
total_params_overall = sum(p.numel() for p in state_dict.values())
print("Total parameters in the model:", total_params_overall)
# Iterate through the keys of the state dictionary
for key in state_dict:
    param = state_dict[key]
    layer_name = key.split('.')[0]  # Extract the layer name
    if layer_name not in num_params_per_layer:
        num_params_per_layer[layer_name] = 0
    num_params_per_layer[layer_name] += param.numel()  # Count the number of parameters
    
    print("Layer:", layer_name)
    print("Key:", key)
    print("Parameter shape:", param.shape)
    print("Parameter dtype:", param.dtype)
    print("Parameter device:", param.device)
    print("Number of parameters:", param.numel())
    # Update the total number of parameters
    total_params += param.numel()

# Print the total number of parameters for each layer
print("\nTotal number of parameters for each layer:")
for layer, num_params in num_params_per_layer.items():
    print(f"{layer}: {num_params}")
# Print the total number of parameters across all layers
print("\nTotal number of parameters across all layers:", total_params)
```
